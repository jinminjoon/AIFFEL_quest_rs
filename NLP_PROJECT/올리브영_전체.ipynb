```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. txt ì»¬ëŸ¼ ì¶”ì¶œ
texts = data['clean_txt'].astype(str)  # í˜¹ì‹œ ëª¨ë¥¼ ê²°ì¸¡ì¹˜/ë¹„ë¬¸ìì—´ ë°©ì§€


# 2. TF-IDF ë²¡í„°ë¼ì´ì € ì´ˆê¸°í™” ë° í•™ìŠµ
# 2. TF-IDF ë²¡í„°ë¼ì´ì € ì„¤ì •
tfidf_vectorizer = TfidfVectorizer(
    max_features=1000,         # ìƒìœ„ 1000ê°œ ë‹¨ì–´ë§Œ ì‚¬ìš©
    stop_words='english',      # ì˜ì–´ ë¶ˆìš©ì–´ ì œê±°
    ngram_range=(2,2),        # unigram + bigram
    min_df=4                   # ìµœì†Œ 2ê°œ ë¬¸ì„œì— ë“±ì¥í•œ ë‹¨ì–´ë§Œ ì‚¬ìš©
)

# 3. ë³€í™˜
tfidf_matrix = tfidf_vectorizer.fit_transform(texts)

# 4. ê²°ê³¼ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
# ë‹¨ì–´ë³„ í‰ê·  TF-IDF ì ìˆ˜ ê³„ì‚°
word_scores = tfidf_df.mean(axis=0).sort_values(ascending=False)

# DataFrameìœ¼ë¡œ ë³€í™˜
top_words_df = word_scores.reset_index()
top_words_df.columns = ['word', 'tfidf_score']

# ìƒìœ„ ë‹¨ì–´ í™•ì¸
print(top_words_df.head(30))
```

```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib import font_manager as fm

# 1. í•œê¸€ í°íŠ¸ ì„¤ì •
font_path = '/content/malgun.ttf'
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['axes.unicode_minus'] = False

# 2. ë„¤íŠ¸ì›Œí¬ ê·¸ë˜í”„ ìƒì„±
G = nx.Graph()
top_n = 30

for row in top_words_df.head(top_n).itertuples():
    words = row.word.split()
    if len(words) == 2:  # âœ… 2-gramë§Œ ì‚¬ìš©
        G.add_edge(words[0], words[1])

# 3. ì‹œê°í™”
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.8)

nx.draw_networkx_nodes(G, pos, node_color='skyblue', node_size=1500)
nx.draw_networkx_edges(G, pos, width=2)

# í•œê¸€ ë¼ë²¨ ìˆ˜ë™ ì¶œë ¥
for node, (x, y) in pos.items():
    plt.text(x, y, node, fontsize=12, fontweight='bold',
             fontproperties=font_prop, ha='center', va='center')

# ì œëª©
plt.title("2-gram TF-IDF ê¸°ë°˜ ë‹¨ì–´ ë„¤íŠ¸ì›Œí¬", fontsize=16, fontproperties=font_prop)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
from afinn import Afinn

# AFINN  ê°ì²´ ìƒì„±
afinn = Afinn()

def safe_text_sentiment_score(text):
    if pd.isnull(text):
        return 0.0
    words = text.split()
    scores = [afinn.score(word) for word in words]
    return sum(scores) / len(scores) if scores else 0.0

data['afinn_score'] = data['translated_txt_clean'].apply(safe_text_sentiment_score)
```
